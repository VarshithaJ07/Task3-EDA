```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
%matplotlib inline
```

```python
from sklearn.datasets import load_iris
iris_raw = load_iris()

iris = pd.DataFrame(iris_raw.data, columns=iris_raw.feature_names)
iris['species'] = iris_raw.target
iris.head()
```

```python
iris['species'] = iris['species'].map({0:"setosa", 1:"versicolor", 2:"virginica"})
```

```python
iris.shape
iris.info()
iris.describe()
iris.isnull().sum()
```

```python
#Numerical distribution
iris.hist(figsize=(10,8), bins=15)
plt.suptitle("Distribution of Iris Features")
plt.show()
```

```python
#Outlier Detection
plt.figure(figsize=(10,6))
sns.boxplot(data=iris.drop(columns='species'))
plt.title("Outlier Detection Using Boxplots")
plt.show()
```

```python
#Species wise comaprision
plt.figure(figsize=(10,6))
sns.boxplot(x='species', y='petal length (cm)', data=iris)
plt.title("Petal Length by Species")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='species', y='petal length (cm)', data=iris)
plt.title("Petal Length by Species")
plt.show()
```

```python
sns.pairplot(iris, hue='species')
plt.suptitle("Pairwise Feature Relationships", y=1.02)
plt.show()
```

```python
numeric_iris = iris.select_dtypes(include=['float64', 'int64'])
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(numeric_iris.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x='petal length (cm)', y='petal width (cm)', hue='species', data=iris)
plt.title("Petal length vs Petal width by Species")
plt.show()
```

#Insights
 Iris dataset contains 150 samples and 3 balanced species.
 No missing values or categorical errors were found.
 Histograms show petal features have lower variance than sepal features.
 Boxplots reveal that Setosa has significantly smaller petal dimensions.
 Pairplot shows Setosa is linearly separable from other species.
 Correlation heatmap shows strong correlation between petal length & width.


#Conclusion
EDA reveals that the Iris dataset is clean, balanced, and highly separable.
Petal measurements act as strong discriminators among species, particularly for Setosa.
This explains why simple models (like logistic regression or k-NN) achieve high accuracy on this dataset.
